```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
X, y, true_coef = make_regression(
    n_samples=100,
    n_features=1,
    n_informative=1,
    noise=10,
    coef=True,
    random_state=0
)
```

```python
# 2. Scale X (experience) to realistic range: 0 to 20 years
X = np.interp(X, (X.min(), X.max()), (0.0, 20.0))
```

```python
# 3. Scale y (salary) to realistic range: ₹20,000 to ₹1,50,000
y = np.interp(y, (y.min(), y.max()), (20000, 150000))
```

```python
# 4. Train Linear Regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# 5. Predict on the same data (for visualization)
y_pred = model.predict(X)
```

```python
# 6. Print model parameters
print(f"Intercept (base salary) : ₹{model.intercept_:.2f}")
print(f"Slope (₹ per year)      : ₹{model.coef_[0]:.2f}")
```

```python
# 7. Visualization
plt.figure(figsize=(9, 6))
plt.scatter(X, y, color='blue', alpha=0.6, label='Actual Data (Training)')
plt.plot(X, y_pred, color='red', linewidth=2.5, label='Linear Regression Fit')
plt.xlabel("Years of Experience")
plt.ylabel("Salary (₹)")
plt.title("Experience vs Salary – Linear Regression")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 8. Example prediction (outside training range)
experience_new = np.array([[22.0]])  # 22 years
predicted_salary = model.predict(experience_new)
print(f"Predicted salary for 22 years of experience: ₹{predicted_salary[0]:.2f}")
```

```python
# Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# 1. Simulate data: y = 10 + 7x + e
n_samples = 100
X = np.random.rand(n_samples, 1)          # X in [0, 1)
error = np.random.randn(n_samples, 1) * 1.5  # Gaussian noise (std ≈ 1.5)
y = 10 + 7 * X + error

# 2. Train Linear Regression model
model = LinearRegression()
model.fit(X, y)

# 3. Predict
y_pred = model.predict(X)

# 4. Print recovered parameters (should be close to 10 and 7)
print(f"Intercept (b0) : {model.intercept_[0]:.4f}  (true = 10)")
print(f"Slope (b1)     : {model.coef_[0][0]:.4f}     (true = 7)")

# 5. Visualization
plt.figure(figsize=(9, 6))
plt.scatter(X, y, color='teal', alpha=0.7, label='Noisy Data (y = 10 + 7x + e)')
plt.plot(X, y_pred, color='darkorange', linewidth=2.8, label='Fitted Line')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simulated Linear Relationship with Noise + Regression Fit")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
